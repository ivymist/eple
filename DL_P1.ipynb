```python
import tensorflow as tf
```

```python
import numpy as np
```

```python
m_a = tf.constant([[1., 2.], [3., 4.]], dtype=tf.float32)
m_b = tf.constant([[5., 6.], [7., 8.]], dtype=tf.float32)
```

```python
print(f"Matrix A:\n{m_a.numpy()}")
print(f"Matrix B:\n{m_b.numpy()}")
```

```python
product = tf.linalg.matmul(m_a, m_b)
```

```python
print(f"\nProduct of A * B:\n{product.numpy()}")
```

```python
symmetric_matrix = tf.constant([[4., 1.], [1., 3.]], dtype=tf.float32)
```

```python
eigenvalues, eigenvectors = tf.linalg.eigh(symmetric_matrix)
```

```python
print(f"\nEigenvalues:\n{eigenvalues.numpy()}")
```

```python
print(f"\nEigenvectors (columns):\n{eigenvectors.numpy()}")
```
